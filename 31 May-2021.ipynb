# leap year

```python
range1=int(input("enter the starting range:"))
range2=int(input("enter the ending range:"))
for i in range(range1,range2,1):
    if i%400==0 or i%4==0 and i%100!=0:
        print(i)
```

```python
for travel in range(0,101,1):
    if travel==50:
        break
    elif travel%2==0:
        continue
    print(travel,end='\t')
```
